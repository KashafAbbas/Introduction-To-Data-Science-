```python
df = pd.read_csv("/kaggle/input/employee-performance-and-salary-dataset/Employe_Performance_dataset.csv")
df.head(10)
```

```python
# Question 2: Column names and data types
print("\n# Question 2: Column names and data types")
print(df.dtypes)
```

```python
# Question 3: Number of rows and columns
print("\n# Question 3: Shape of the dataset")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
# Question 4: Select Name, Age, and Salary columns
print("\n# Question 4: Name, Age, and Salary columns")
print(df[['Name', 'Age', 'Salary']])
```

```python
# Question 5: First 5 rows using iloc
print("\n# Question 5: First 5 rows using iloc")
print(df.iloc[:5])
```

```python
# Question 6: Rows where Department is "Sales"
print("\n# Question 6: Rows where Department is 'Sales'")
print(df.loc[df['Department'] == 'Sales'])
```

```python
# Question 7: Salary > 8000 and Status is "Active"
print("\n# Question 7: Salary > 8000 and Status is 'Active'")
print(df.loc[(df['Salary'] > 8000) & (df['Status'] == 'Active')])
```

```python
# Question 8: Employees with >5 years and Performance Score >= 4
print("\n# Question 8: Employees with >5 years and Performance Score >= 4")
print(df.loc[(df['Experience'] > 5) & (df['Performance Score'] >= 4)])
```

```python
# Question 9: Every alternate row
print("\n# Question 9: Every alternate row")
print(df.iloc[::2])
```

```python
# Question 10: Sales employees with top 10% salaries
print("\n# Question 10: Sales employees with top 10% salaries")
top_10_percent_salary = df['Salary'].quantile(0.9)
print(df.loc[(df['Department'] == 'Sales') & (df['Salary'] >= top_10_percent_salary)])
```

```python
# Question 11: Replace "Inactive" with "Retired" in Status
print("\n# Question 11: Replace 'Inactive' with 'Retired'")
df['Status'] = df['Status'].replace('Inactive', 'Retired')
print(df['Status'])
```

```python
# Question 12: Add a Bonus column (10% of Salary)
print("\n# Question 12: Add Bonus column")
df['Bonus'] = df['Salary'] * 0.1
print(df[['Salary', 'Bonus']])
```

```python
# Question 13: Update Performance Score to 5 for NaN in IT department
print("\n# Question 13: Update Performance Score in IT department")
df.loc[(df['Department'] == 'IT') & (df['Performance Score'].isna()), 'Performance Score'] = 5
print(df.loc[df['Department'] == 'IT', 'Performance Score'])
```

```python
# Question 14: Modify Experience < 2 to 2
print("\n# Question 14: Modify Experience < 2 to 2")
df.loc[df['Experience'] < 2, 'Experience'] = 2
print(df['Experience'])
```

```python
# Question 15: Normalize Salary between 0 and 1
print("\n# Question 15: Normalize Salary")
df['Salary Normalized'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
print(df[['Salary', 'Salary Normalized']])
```

```python
# Question 16: Replace Location with "Remote" for Performance Score < 3
print("\n# Question 16: Replace Location with 'Remote' for Performance Score < 3")
df.loc[df['Performance Score'] < 3, 'Location'] = 'Remote'
print(df[['Performance Score', 'Location']])
```

```python
# Question 17: Create Experience Level column
print("\n# Question 17: Create Experience Level column")
def experience_level(exp):
    if exp < 5:
        return 'Junior'
    elif 5 <= exp < 10:
        return 'Mid'
    else:
        return 'Senior'
df['Experience Level'] = df['Experience'].apply(experience_level)
print(df[['Experience', 'Experience Level']])
```

```python
# Question 18: Add a new row using concat
print("\n# Question 18: Add a new row")
new_row = pd.DataFrame([{'Name': 'New Employee', 'Age': 30, 'Salary': 5000, 'Department': 'HR', 
                         'Experience': 3, 'Performance Score': 4, 'Status': 'Active', 'Location': 'Remote'}])
df = pd.concat([df, new_row], ignore_index=True)
print(df.tail())
```

```python
# Question 19: Delete the column Session
print("\n# Question 19: Delete the column 'Session'")
if 'Session' in df.columns:
    df = df.drop(columns=['Session'])
print(df.columns)
```

```python
# Question 20: Delete rows where Age > 60
print("\n# Question 20: Delete rows where Age > 60")
df = df[df['Age'] <= 60]
print(df['Age'].max())
```

```python
# Question 21: Add Age Group column
print("\n# Question 21: Add Age Group column")
def age_group(age):
    if age < 30:
        return 'Young'
    elif 30 <= age < 50:
        return 'Middle-aged'
    else:
        return 'Senior'
df['Age Group'] = df['Age'].apply(age_group)
print(df[['Age', 'Age Group']])
```

```python
# Question 22: Identify columns with missing values
print("\n# Question 22: Identify columns with missing values")
missing_columns = df.isna().sum()
print(missing_columns[missing_columns > 0])
```

```python
# Question 23: Fill missing values in Performance Score with the average score
print("\n# Question 23: Fill missing Performance Score values with the average")
average_score = df['Performance Score'].mean()
df['Performance Score'].fillna(average_score, inplace=True)
print(df['Performance Score'].isna().sum())
```

```python
# Question 24: Drop all rows with missing values
print("\n# Question 24: Drop rows with missing values")
df = df.dropna()
print(df.isna().sum().sum())
```

```python
# Question 25: Replace missing Performance Score with median grouped by Department
print("\n# Question 25: Replace missing Performance Score with median grouped by Department")
median_scores = df.groupby('Department')['Performance Score'].transform('median')
df['Performance Score'] = df['Performance Score'].fillna(median_scores)
print(df['Performance Score'])
```

```python
# Question 26: Identify employees in IT with Age > IT average
print("\n# Question 26: Employees in IT with Age > IT average")
it_average_age = df.loc[df['Department'] == 'IT', 'Age'].mean()
print(df.loc[(df['Department'] == 'IT') & (df['Age'] > it_average_age)])
```

```python
# Question 27: Remove duplicate Name and Joining Date rows
print("\n# Question 27: Remove duplicate Name and Joining Date rows")
df = df.drop_duplicates(subset=['Name', 'Joining Date'], keep='first')
print(df)
```

```python
# Question 28: Remove rows where Experience > Age - 18
print("\n# Question 28: Remove rows with inconsistent Experience")
df = df[df['Experience'] <= (df['Age'] - 18)]
print(df)
```

```python
# Question 29: Sort by Salary in descending order
print("\n# Question 29: Sort by Salary in descending order")
df = df.sort_values(by='Salary', ascending=False)
print(df[['Name', 'Salary']])
```

```python
# Question 30: Sort by Department and Joining Date
print("\n# Question 30: Sort by Department and Joining Date")
df = df.sort_values(by=['Department', 'Joining Date'])
print(df[['Department', 'Joining Date']])
```

```python
# Question 31: Rank employees by Salary within Department
print("\n Question 31: Rank employees by Salary within Department")
df['Salary Rank'] = df.groupby('Department')['Salary'].rank(ascending=False)
print(df[['Department', 'Salary', 'Salary Rank']])
```

```python
# Question 32: Sort by Experience, break ties with Performance Score
print("\n# Question 32: Sort by Experience and Performance Score")
df = df.sort_values(by=['Experience', 'Performance Score'], ascending=[False, True])
print(df[['Experience', 'Performance Score']])
```

```python
# Question 33: Create new DataFrame with HR employees and Salary > 5000
print("\n# Question 33: HR employees with Salary > 5000")
hr_high_salary = df.loc[(df['Department'] == 'HR') & (df['Salary'] > 5000)]
print(hr_high_salary)
```

```python
# Question 34: Filter out rows where Location is Chicago
print("\n# Question 34: Filter out rows where Location is 'Chicago'")
df = df[df['Location'] != 'Chicago']
print(df['Location'].unique())
```

```python
# Question 35: Delete rows where Performance Score < 2
print("\n# Question 35: Delete rows where Performance Score < 2")
df = df[df['Performance Score'] >= 2]
print(df['Performance Score'].min())
```

```python
# Question 36: Average Salary for each Department
print("\n# Question 36: Average Salary for each Department")
average_salary = df.groupby('Department')['Salary'].mean()
print(average_salary)
```

```python
# Question 37: Total Salary grouped by Status
print("\n# Question 37: Total Salary grouped by Status")
total_salary_status = df.groupby('Status')['Salary'].sum()
print(total_salary_status)
```

```python
# Question 38: Maximum Performance Score by Location
print("\n# Question 38: Maximum Performance Score by Location")
max_perf_location = df.groupby('Location')['Performance Score'].max()
print(max_perf_location)
```

```python
# Question 39: Total Salary by Joining Year
print("\n# Question 39: Total Salary by Joining Year")
df['Joining Year'] = pd.to_datetime(df['Joining Date']).dt.year
total_salary_year = df.groupby('Joining Year')['Salary'].sum()
print(total_salary_year)
```

```python
# Question 40: Department with highest average Performance Score
print("\n# Question 40: Department with highest average Performance Score")
highest_perf_department = df.groupby('Department')['Performance Score'].mean().idxmax()
print(highest_perf_department)
```

```python
# Question 41: Count employees by Location and Session
print("\n# Question 41: Count employees by Location and Session")
if 'Session' in df.columns:
    location_session_count = df.groupby(['Location', 'Session']).size()
    print(location_session_count)
else:
    print("The 'Session' column does not exist in the dataset.")
```

```python
# Question 42: Salary summary stats by Status
print("\n# Question 42: Salary summary stats by Status")
status_salary_stats = df.groupby('Status')['Salary'].agg(['sum', 'mean', 'std'])
print(status_salary_stats)
```

```python
# Question 43: Create Monthly Salary column
print("\n# Question 43: Create Monthly Salary column")
df['Monthly Salary'] = df['Salary'] / 12
print(df[['Salary', 'Monthly Salary']])
```

```python
# Question 44: Convert all strings to uppercase
print("\n# Question 44: Convert all strings to uppercase")
df = df.applymap(lambda x: x.upper() if isinstance(x, str) else x)
print(df)
```

```python
# Question 45: Inner join with Department and Budget dataset
print("\n# Question 45: Inner join with Department and Budget")
dep_budget = pd.DataFrame({'Department': ['HR', 'IT', 'Sales'], 'Budget': [50000, 70000, 80000]})
inner_join = pd.merge(df, dep_budget, on='Department', how='inner')
print(inner_join)
```

```python
# Question 46: Merge with ID and Projects dataset
print("\n# Question 46: Merge with ID and Projects")
emp_projects = pd.DataFrame({'ID': [1, 2, 3], 'Recent Projects': ['Proj1', 'Proj2', 'Proj3']})
merged_df = pd.merge(df, emp_projects, on='ID', how='left')
print(merged_df)
```

```python
# Question 47: Left join with Department and Team Lead
print("\n# Question 47: Left join with Department and Team Lead")
dep_team_lead = pd.DataFrame({'Department': ['HR', 'IT', 'Sales'], 'Team Lead': ['Alice', 'Bob', 'Charlie']})
left_join = pd.merge(df, dep_team_lead, on='Department', how='left')
print(left_join)
```

```python
# Question 48: Merge with Project Count, calculate avg Salary per project
print("\n# Question 48: Merge with Project Count and calculate avg Salary per project")
if 'ID' in df.columns:
    project_count = pd.DataFrame({'ID': [1, 2, 3], 'Project Count': [5, 3, 2]})
    df = pd.merge(df, project_count, on='ID', how='left')
    df['Avg Salary per Project'] = df['Salary'] / df['Project Count']
    print(df[['ID', 'Salary', 'Project Count', 'Avg Salary per Project']])
else:
    print("The 'ID' column does not exist in the dataset.")
```

```python
# Question 49: Percentage increase in Salary by Department average
print("\n# Question 49: Percentage increase in Salary by Department average")
dep_avg_salary = df.groupby('Department')['Salary'].transform('mean')
df['Salary Increase %'] = ((df['Salary'] - dep_avg_salary) / dep_avg_salary) * 100
print(df[['Department', 'Salary', 'Salary Increase %']])
```
