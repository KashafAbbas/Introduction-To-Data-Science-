```python
#part 1
users = [(1, "kash", 25, "USA"), (2, "bob", 32, "Canada"), (3, "sarah", 40, "Canada"), (4, "rory", 28, "USA"),(5, "kash", 25, "UK")]

#filter names
def filterusers(users):
    filterednames = [user[1] for user in users if user[2] > 30 and user[3] in ['USA', 'Canada']]
    return filterednames

filterednames = filterusers(users)
print("Filtered names of users older than 30 from USA or Canada:", filterednames)
print()

#sort names by age and check for duplicate names
def sortbyage(users):
    sortedusers = sorted(users, key=lambda x: x[2], reverse=True)[:10]
    return sortedusers

sortedusers = sortbyage(users)
print("Sorted users by age:", sortedusers)
print()

def duplicatenames(users):
    names = [user[1] for user in users]
    namecounts = {name: names.count(name) for name in names}
    duplicates = [name for name, count in namecounts.items() if count > 1]
    return duplicates

duplicates = duplicatenames(users)
print("Duplicate names in the list of users:", duplicates)
print()

#part 2

trans = [(101, 1, 500, "2024-01-01"), (102, 2, 1000, "2024-01-02"), (103, 1, 750, "2024-01-03")]

#unique users
def unqusr(trans):
    unqusrids = set(tran[1] for tran in trans)
    return len(unqusrids)

unqusrcount = unqusr(trans)
print("Total number of unique users in transactions:", unqusrcount)
print()

#highest transactions
def maxtrans(trans):
    maxtrans = max(trans, key=lambda x: x[2])
    return maxtrans

highesttrans = maxtrans(trans)
print("Transaction with the highest amount:", highesttrans)
print()

#separate lists
def sepids(trans):
    transids = [tran[0] for tran in trans]
    usrids = [tran[1] for tran in trans]
    return transids, usrids

transids, usrids = sepids(trans)
print("Transaction IDs and User IDs separated:", transids, usrids)
print()

#part 3
pagea = {1, 2, 3}
pageb = {2, 3, 4}
pagec = {3, 4, 5}

#users who visited both Page A and Page B.
def commonusers(pagea, pageb):
    return pagea.intersection(pageb)

common = commonusers(pagea, pageb)
print("Users who visited both Page A and Page B:", common)
print()

#users who visited either Page A or Page C, but not both.
def eitherbutnotboth(pagea, pagec):
    return pagea.symmetric_difference(pagec)

eithernotboth = eitherbutnotboth(pagea, pagec)
print("Users who visited either Page A or Page C, but not both:", eithernotboth)
print()

#Update the set with new user IDs.
def updateseta(pagea, newids):
    pagea.update(newids)
    return pagea

#Removes a list of user IDs from the set
def removeusers(pageb, userstoremove):
    pageb.difference_update(userstoremove)
    return pageb

newa = updateseta(pagea, {6, 7})
print("Page A after updating with new user IDs:", newa)
print()

newb = removeusers(pageb, {2})
print("Page B after removing user IDs:", newb)
print()

#part 4

feedback = {1: {'rating': 5, 'comments': 'excellent'}, 2: {'rating': 3, 'comments': 'good'}, 3: {'rating': 4, 'comments': 'great'}}

#Filters out users who rated 4 or higher and stores their user_id and rating
def filterfeedback(feedback):
    filtered = {userid: details['rating'] for userid, details in feedback.items() if details['rating'] >= 4}
    return filtered

filteredfeedback = filterfeedback(feedback)
print("Filtered feedback with rating 4 or higher:", filteredfeedback)
print()

#sort the dictionary of user feedback
def sortfeedback(feedback):
    sortedfeedback = sorted(feedback.items(), key=lambda x: x[1]['rating'], reverse=True)[:5]
    return sortedfeedback

topfeedback = sortfeedback(feedback)
print("Top 5 users based on feedback sorted by rating:", topfeedback)
print()

#Combines feedback from multiple dictionaries.
def combinefeedback(dicts):
    combined = {}
    for feedback in dicts:
        for userid, details in feedback.items():
            if userid in combined:
                combined[userid]['rating'] = max(combined[userid]['rating'], details['rating'])
                combined[userid]['comments'] += " " + details['comments']
            else:
                combined[userid] = details.copy()
    return combined

feedback1 = {1: {'rating': 4, 'comments': 'good'}, 2: {'rating': 5, 'comments': 'excellent'}}
feedback2 = {2: {'rating': 4, 'comments': 'great'}, 3: {'rating': 3, 'comments': 'okay'}}
feedback3 = {1: {'rating': 5, 'comments': 'awesome'}, 3: {'rating': 4, 'comments': 'improved'}}

combinedfeedback = combinefeedback([feedback1, feedback2, feedback3])
print("Combined feedback from multiple dictionaries:", combinedfeedback)
print()

#Use dictionary comprehension to filter ratings greater than 3
def filterratings(feedback):
    filtered = {userid: details['rating'] for userid, details in feedback.items() if details['rating'] > 3}
    return filtered

filteredratings = filterratings(combinedfeedback)
print("Filtered ratings greater than 3:", filteredratings)
```
